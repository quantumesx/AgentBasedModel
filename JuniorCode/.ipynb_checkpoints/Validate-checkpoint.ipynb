```python
with open('FirstGen/Run0669.txt', 'r') as file:
    g_raw = file.readlines()
    g = [int(l.strip()) for l in g_raw]
#print(g)

plt.figure(1, figsize=(12, 3))
plt.subplot(121)
n, bins, patches = plt.hist(x=g, bins=[i for i in range(0, 256, 20)], color='#0504aa', alpha=0.7, rwidth=0.85)
plt.grid(axis='y', alpha=0.75)
plt.ylim(ymax=np.ceil(12))

plt.subplot(122)
plt.plot(range(0, 69), g)
```

```python
def save_first_gen_files(genome, filename):
    with open(filename, 'w') as f:
        for g in genome:
            g_s = [str(i) for i in g]
            f.write(",".join(g_s))
            f.write('\n')
            
def read_first_gen_files(filename):
    with open(filename, 'r') as f:
        first_gen = []
        raw = f.readlines()
        for l in raw:
            g = [int(s) for s in l.strip('\n').split(',')]
            first_gen.append(g)

    return first_gen
```
